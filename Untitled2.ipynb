```python
import folium
from folium import plugins
import pandas as pd
```

```python
df=pd.read_csv('data1.csv')
df.reset_index(drop=True,inplace=True)
df.rename(columns={'Postal code':'Postal_code'},inplace=True)
df=df[df['Borough']!="Not assigned"]
print(df.shape)
df.groupby('Postal_code')
#no empty null values in neighborhood
print(df['Neighborhood'].isna().sum())
df.head()
df_geo=pd.read_csv('Geospatial_Coordinates.csv')

print("df_geo shape:",df_geo.shape)
print("df:",df.shape)
df.sort_values('Postal_code',inplace=True)
df.reset_index(drop=True,inplace=True)
df[['Latitude']]=df_geo[['Latitude']]
df[['Longitude']]=df_geo[['Longitude']]
#check that there are none null values
print(df.isnull().sum())
```

```python
map_tor= folium.Map(location=[43.6532, -79.3832],zoom_start=12)

incidents = plugins.MarkerCluster().add_to(map_tor)
for x, y,z in zip(df.Latitude, df.Longitude,df.Postal_code):
    folium.features.CircleMarker(
        [x, y],
        radius=5, # define how big you want the circle markers to be
        color='yellow',
        fill=True,
        popup=z,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(incidents)

map_tor
```
